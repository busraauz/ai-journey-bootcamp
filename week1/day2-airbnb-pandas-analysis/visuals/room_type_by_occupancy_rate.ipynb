```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("../data/airbnb_clean.csv")
df.head()
```

```python
room_type_occupancy_rate = (1 - df.groupby("room_type")["availability_rate"].mean()).reset_index()
room_type_occupancy_rate
```

```python
fig, ax = plt.subplots()

labels = room_type_occupancy_rate["room_type"].str.capitalize()
rates = room_type_occupancy_rate["availability_rate"]

ax.bar(labels, rates)

ax.set_ylabel('Availability Rates')
ax.set_xlabel('Room Type')

ax.set_title('Room Type By Occupancy Rate')

plt.show()
```
