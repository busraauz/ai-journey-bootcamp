```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_csv("../data/online_retail.csv")
df.head()
```

```python
# Quick Overview
df.describe()
df.isna().sum()
```

```python
# Handling Missing Data, Data Type Conversion
df = df[df["Customer ID"].notna()]
df["Customer ID"] = df["Customer ID"].astype("int64")

df["Description"] = df["Description"].fillna("Unknown")
df
```

```python
# Detect and Extract Outliers
df = df[df.Price > 0]
df = df[df.Quantity > 0]
```

```python
# Create New Informational Columns
df["Total Price"] = df["Price"] * df["Quantity"]
date_time = pd.to_datetime(df['InvoiceDate'])
df["InvoiceDate"] = date_time
df["InvoiceYear"] = date_time.dt.year
df["InvoiceMonth"] = date_time.dt.month
df
```

```python
df.to_csv("../data/online_retail_clean.csv", index=False)
```
